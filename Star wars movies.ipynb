```python
import numpy as np
yes_no1={
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
for i in star_wars.columns[3:9]:
    star_wars[i]=star_wars[i].map(yes_no1)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
star_wars.head(5)
```

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
avg1=star_wars["ranking_1"].mean()
avg2=star_wars["ranking_2"].mean()
avg3=star_wars["ranking_3"].mean()
avg4=star_wars["ranking_4"].mean()
avg5=star_wars["ranking_5"].mean()
avg6=star_wars["ranking_6"].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
plt.xlabel("ranking")
plt.ylabel("average_score")
plt.show()
```

Looking at the mean of all the movies, we can see that people appreciate more the fourth movie, followed by the the fifth and the third.

```python
plt.bar(range(6),star_wars[star_wars.columns[3:9]].sum())
plt.xlabel("movie")
plt.ylabel("number of viewers")
plt.show()
```

The most seen movies are the fifth and the fourth. This is related to the ranking because also the fourth and fifth have the highest ranking
